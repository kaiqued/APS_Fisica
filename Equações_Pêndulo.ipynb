```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

%matplotlib inline
```

```python
from scipy.integrate import odeint
g=9.8
theta=radians(90)
L=2

def equidif(Soluções,t):
    Theta=Soluções[0]
    w =Soluções[1]
    dThetadt = w
    dwdt = -(g*sin(Theta))/L
    return [dThetadt,dwdt]
dt=0.001

listaTempo = np.arange(0,4, dt)
cond_inicial = [theta,0]
```

```python
solucao = odeint(equidif, cond_inicial, listaTempo)
X=[]
Y=[]
for i in solucao[:,0]:
    x= L*cos(i)
    y= L*sin(i)
    X.append(x)
    Y.append(y)
     
plt.plot(Y,X)
plt.title('Posição Do Pêndulo')
plt.grid(True)
plt.axis([min(Y),max(Y),max(X),min(X)])
plt.axis('equal')
plt.show()

plt.title('Velocidade ')
plt.plot(listaTempo,solucao[:,0])
plt.grid(True)
plt.show()
```
